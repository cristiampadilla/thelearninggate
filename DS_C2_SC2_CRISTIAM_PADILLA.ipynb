```python
import numpy as nm
import pandas as pd

print("A. Extraer la información del archivo  ")
df=pd.read_csv("/content/country_vaccinations.csv")

print(df.head(10))
print("------------------------------------------------------------------------")

print("B. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.")
df['date'] = df['date'].astype('datetime64')
print(df.dtypes)

print("------------------------------------------------------------------------")
print("C. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).")

dfC = df.groupby('vaccines')['daily_vaccinations'].sum()
print(dfC)

print("------------------------------------------------------------------------")
print("D. Obtener la cantidad de vacunas aplicadas en todo el mundo. ")

print(df['daily_vaccinations'].sum())

print("------------------------------------------------------------------------")
print("E. Calcular el promedio de vacunas aplicadas por país diariamente.")
dfE = df.groupby('country')['daily_vaccinations'].mean()
print(dfE)

print("------------------------------------------------------------------------")
print("F. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo")
dfF = df.loc[df['date'] == '29/01/21']
print(dfF['daily_vaccinations'].sum())

print("------------------------------------------------------------------------")
print("G. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.")
conDiferencias = df.copy()
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']

print(conDiferencias.head(98) )

print("------------------------------------------------------------------------")
print("H. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua")

print(df['date'].min())
print(df['date'].max())
print(df['date'].max() - df['date'].min())

print("------------------------------------------------------------------------")
print("i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).")

conCantidad = df.copy()
conCantidad['canVac'] = conCantidad['vaccines'].str.count(',').add(1)

print(conCantidad.head(100) )

print("------------------------------------------------------------------------")
print("J. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020")

antes20 = df.loc[df['date'] < '20/12/2020']
print(antes20.head(100))

print("------------------------------------------------------------------------")
print("K. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer")
pfizer = df.loc[df['vaccines'].str.contains('Pfizer')]

print(pfizer)

print("------------------------------------------------------------------------")
print("L. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente")

with pd.ExcelWriter("/content/resultadosReto.xlsx") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias")
    conCantidad.to_excel(writer, sheet_name="conCantidad")
    antes20.to_excel(writer, sheet_name="antes20")
    pfizer.to_excel(writer, sheet_name="pfizer")
```
